```python
# finding the null values percentage
df.isnull().sum()* 100 /len(df)
```

```python
# As we know if a column has more than 60 % null values it is a fully to work with it
# so here we drop the deck attribute
df.drop(columns={'deck'}, inplace=True)
```

```python
df.head()
```

```python
# now replacing the null values

# we usally replace it by the mean or the median of it if its a categorical then mode

# here we are replacing the age column null values by the median value

df['age']= df['age'].fillna(df['age'].median())
```

```python
df.isnull().sum()
```

```python
# lets us check the null values in the embarked and embarked town
df.sort_values(by='embarked', na_position='first')
```

We can see that null value for embarked and embarked_town town is the same so we don't have to drop separates we can do it dealing with one of attribute

```python
# now for embarked and embarked town we are droping the null values rows

df.dropna(inplace=True)
```
