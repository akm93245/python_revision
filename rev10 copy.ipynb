```python
# facetgrid makinf of plot for every unique value in the column
graph = sns.FacetGrid(df,hue='survived' , col='pclass', margin_titles=True , palette={1:'seagreen',0:'grey'})
```

```python
graph = graph.map(plt.scatter, 'fare' ,'age' ,edgecolor='w').add_legend()

plt.show()
```

```python
# Creating a FacetGrid plot for every unique value in the 'pclass' column
graph = sns.FacetGrid(df, hue='survived', col='pclass', margin_titles=True, palette={1: 'seagreen', 0: 'grey'})

# Mapping the scatter plot to the FacetGrid
graph = graph.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()

# Display the plot
plt.show()
```

```python
# creating of facet gid ( between the uniques of two variables)
graph = sns.FacetGrid(data=df, # the data 
                      hue='survived', # the colour column
                      col='pclass', # Column which you wanna make the facetgrid of
                      margin_titles=True,
                      palette={1:"blue",0:'red'}
                      )
graph= graph.map(plt.scatter, # the plot you wanna make
                 'fare', # the x - axis on whih p class vars will be made 
                 'age', # the y-axis 
                 edgecolor='w'
                 ).add_legend()
plt.show()
```

```python
# creating the strip plot
# the plot which is made of both bars and scatter
sns.stripplot(data=df, x='sex', y='age', hue='pclass')
```

```python
# creating voilin plot
sns.violinplot(data=df, x='survived', y='fare', hue='pclass' )
```
